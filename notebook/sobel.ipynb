```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("../data/image_1.png")[:, :, ::-1]

plt.imshow(img) 

def apply_sobel_filter(image):

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=25)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=25)
    
    return sobelx, sobely

sobelx, sobely = apply_sobel_filter(img)

plt.subplot(1, 2, 1)
plt.imshow(sobelx, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(sobely, cmap='gray')
plt.show()

plt.imshow(np.sqrt(sobelx**2 + sobely**2), cmap='gray')
plt.show()
```
