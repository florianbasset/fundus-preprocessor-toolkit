```python
import cv2

import matplotlib.pyplot as plt

img = cv2.imread("../data/image_1.png")[:, :, ::-1]

plt.imshow(img) 

def apply_sobel_filter(image):

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=25)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=25)
    
    return sobelx, sobely

plt.imshow(apply_sobel_filter(img)[0], cmap='gray')
plt.show()
plt.imshow(apply_sobel_filter(img)[1], cmap='gray')
plt.show()
```
