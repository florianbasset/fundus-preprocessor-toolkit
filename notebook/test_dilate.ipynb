```python
import cv2

import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../data/image_1.png")[:, :, ::-1] #comme la fonction cv2.cvtColor
# H x W x C
roi = img[:, :, 0] > 10
plt.show(plt.imshow(roi))
img_dilated = cv2.dilate(img, cv2.getStructuringElement(cv2.MORPH_DILATE, (175, 175)))
plt.show(plt.imshow(img_dilated))
img_dilated[roi] = img[roi]
plt.show(plt.imshow(img_dilated))
```
