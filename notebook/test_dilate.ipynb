```python
import cv2

import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../data/image_1.png")[:, :, ::-1]
# H x W x C
roi = img[:, :, 0] > 10
img_dilated = cv2.dilate(img, cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (40, 40)))

img_dilated[roi] = img[roi]
plt.show(plt.imshow(img_dilated))
```
